### Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

### Generate Synthetic Data 

```python
# Generate random data
np.random.seed(0)
size = np.random.randint(1000, 3500, 100)  # House sizes in square feet
bedrooms = np.random.randint(1, 5, 100)    # Number of bedrooms

# Combine features into a single array
features = np.column_stack((size, bedrooms))
```

### Apply K-Means Clustering

```python
# Initialize the K-Means model
kmeans = KMeans(n_clusters=3, random_state=0)

# Fit the model to the data
kmeans.fit(features)

# Predict cluster labels for the data
clusters = kmeans.predict(features)

# Get the cluster centers
centers = kmeans.cluster_centers_
```

### Plot the Clustering Results

```python
# Create a scatter plot of the data, colored by cluster
plt.figure(figsize=(10, 6))
scatter = plt.scatter(features[:, 0], features[:, 1], c=clusters, cmap='viridis', marker='o', s=100, alpha=0.6)

# Plot cluster centers
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x', label='Cluster Centers')

# Add labels and title
plt.xlabel("Size (sq ft)", fontsize=14)
plt.ylabel("Number of Bedrooms", fontsize=14)
plt.title("K-Means Clustering of Houses", fontsize=16)

# Add a legend
legend1 = plt.legend(*scatter.legend_elements(), title="Clusters", fontsize=12)
plt.legend(loc='upper right', fontsize=12)
plt.gca().add_artist(legend1)

# Annotate cluster centers
for i, (x, y) in enumerate(centers):
    plt.text(x, y, f"Center {i+1}", fontsize=12, ha='right')

plt.grid(True)
plt.show()
```

### Make Predictions and Evaluate the Model

```python
# New data point for prediction: [size, bedrooms]
new_data = np.array([[2200, 3]])

# Predict the cluster for the new data point
new_cluster = kmeans.predict(new_data)

# Output the prediction
print(f"New house with size {new_data[0][0]} sq ft and {new_data[0][1]} bedrooms is predicted to be in Cluster {new_cluster[0] + 1}")

# Visualize the new data point in the plot
plt.figure(figsize=(10, 6))
scatter = plt.scatter(features[:, 0], features[:, 1], c=clusters, cmap='viridis', marker='o', s=100, alpha=0.6)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='x', label='Cluster Centers')

# Plot the new data point
plt.scatter(new_data[:, 0], new_data[:, 1], c='blue', s=300, alpha=0.75, marker='*', label='New Data Point')

# Add labels and title
plt.xlabel("Size (sq ft)", fontsize=14)
plt.ylabel("Number of Bedrooms", fontsize=14)
plt.title("K-Means Clustering of Houses", fontsize=16)

# Add a legend
legend1 = plt.legend(*scatter.legend_elements(), title="Clusters", fontsize=12)
plt.legend(loc='upper right', fontsize=12)
plt.gca().add_artist(legend1)

# Annotate cluster centers
for i, (x, y) in enumerate(centers):
    plt.text(x, y, f"Center {i+1}", fontsize=12, ha='right')

# Annotate the new data point
plt.text(new_data[0][0], new_data[0][1], "New Data Point", fontsize=12, ha='left', color='blue')

plt.grid(True)
plt.show()
```
